```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

Carl's coin is secretly biased with p(heads) = 0.4. We think it's likely that the coin is fair, but we are not sure. We place a prior categorical distribution over the true parameter of the coin. Our prior is gives 0.8 to p(heads) = .5, 0.1 to p(heads) = 0.4, and 0.1 to p(heads) = 0.6. Given our prior and our observations, we try to determine how likely each of three parameters are.

- posterior is proportional to likelihood * prior
- If P = p(heads), H = num_heads, and N = num_flips, our likelihood function is:
$$\theta^{N_1}(1-\theta)^{N-N_1}$$

```python
params  = np.array([0.4, 0.5, 0.6])
priors = np.array([0.1, 0.8, 0.1])
heads = 0
secret_truth = 0.4

for i in range(1,31):
    # flip a coin
    r = random.uniform(0, 1)
    if r < secret_truth:
        heads += 1
        print "Flipped Heads"
    else:
        print "Flipped Tails"
    
    print "Heads / Total Flips:",heads,'/',i
    
    likelihood = np.multiply(np.power(params,heads),np.power(1-params,i-heads))
    posterior = np.multiply(likelihood,priors)
    
    # normalization
    posterior = posterior / float(sum(posterior))
    
    # plotting
    plt.bar(params,posterior,width=0.05)
    plt.xlim([0,1])
    plt.ylim([0,1])
    plt.show()
```
